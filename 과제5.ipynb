### 다형성이란 무엇인가
상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능

### 다형성의 특성을 보여주는 예제

```python
class Color :
    def applyColor(self) :
        print("non-defined")
class Red(Color) :
    def applyColor(self) :
        print("Red is applied")
class Blue(Color) :
    def applyColor(self) :
        print("Blue is applied")
class Silver(Color) :
    def applyColor(self) :
        print("Silver is applied")
class Brown(Color) :
    pass

for ele in Red(), Blue(), Silver(), Brown() :
    ele.applyColor()
```

어떠한 물건에 대한 색을 정할때, for 문을 이용해 모든 클래스마다 있는 메소드 applyColor에 대한 접근을 쉽게한다. 또한 Brown은 applyColor 메소드가 정의되어있지 않지만, Color 클래스를 상속받았기 때문에 Brown 클래스의 applyColor 메소드가 호출되면 Color 클래스에 있는 applyColor 메소드가 호출된다.

## 3번

```python
class Counter :
    def __init__(self, count, step = 1) :
        self.count = count
        self.step = step
    def __str__(self) :
        return "[Count(step: %d)]%d " % (self.step, self.count)
    def incr(self) :
        self.count += self.step
    def __call__(self) :
        self.incr()
    def __add__(self, other) :
        self.count += other
        return self
    def __sub__(self, other) :
        self.count -= other
        return self
    def __gt__(self, other) :
        if self.count > other :
            return True
        else :
            return False
    def __lt__(self, other) :
        if self.count < other:
            return True
        else :
            return False
    def __eq__(self, other) :
        if self.count == other :
            return True
        else :
            return False
    def __ne__(self, other):
        if self.count != other :
            return True
        else :
            return False
        
c = Counter(10)
d = Counter(20,2)
print(c)
print(d)

print()

c.incr()
d.incr()
print(c)
print(d)

print()

c()
d()
print(c)
print(d)

print()

c = c+5
d = d+5
print(c)
print(d)

print()

print(c > 10)
print(d > 50)
print(c < 5)
print(d < 30)
print(c == 20)
print(d != 20)
```

\_\_init\_\_ : 생성자  
\_\_str\_\_ : print(), str()에 대응, [Count(step: self.step)]self.count 형식으로 출력. 포매팅방식 사용함  
\_\_call\_\_ : 인스턴스 호출에 해당  
\_\_add\_\_, \_\_sub\_\_ : +, - 에 해당  
\_\_gt\_\_, \_\_lt\_\_, \_\_eq\_\_, \_\_ne\_\_ : >, <, =, != 에 해당


## 문제4

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

\_\_init\_\_  

생성자메소드. MySet 클래스의 인스턴스가 만들어질 때 자동으로 호출된다( ex. a = MySet([1,2,3]) ). 이 때 인자로 설정되어있는 self는 해당 클래스의 인스턴스(a)이고, l은 인자로 넘겨받은 리스트[1,2,3] 이다.  
self는 클래스 명으로 함수 호출(Myset.\_\_init\_\_())을 제외하면 자동으로 넘겨지므로 따로 추가할 필요가 없다. 생성자에서는 인자로 넘겨받은 l리스트의 각각의 원소에 for문으로 접근해 MySet의 리스트객체에 append 시킨다. Myset클래스는 list클래스를 상속받고 있으므로 클래스내부에 리스트객체가 존재한다.  
마지막으로 MySet 클래스 내부에 정의되어있는 eliminate_duplicate() 메소드에 self를 인자로 넘겨준다.

\_\_str\_\_  

print() 또는 str() 메소드에 대응하는 함수이다. print(객체명) 혹은 str(객체명) 으로 사용시에 호출된다.  또한 생성자보다 먼저 호출되어 MySet에 대응하는 문자열을 정의한다.
result라는 문자열객체는 "MySet: {" 으로 시작한다. 생성자에서 eliminate_duplicate() 메소드를 통해 self 리스트객체가 만들어졌고, 이 메소드 내에서 for문을 통해 self의 각 원소(e)들에 접근하여 result 문자열에 e를 문자열로 변환(str(e))한 것을 추가한다.  
마지막으로 result는 0번째 인덱스부터 맨 마지막 인덱스에서 2를 뺀 index ( [0:len(result)-2] ) 에 '}' 를 더해준 문자열을 저장한다. 2를 뺀 이유는 ' ', ',' 이 두 글자를 없애주기 위함.

eliminate_duplicate  

위 두 메소드들과는 다르게 예약된 이름이 아닌 사용자가 정의한 메소드이다.  또한 정적메소드(static method) 이다.
인자로는 생성자에서 self를 넘겨주어 리스트(l)로 받는다. 이 메소드의 역할은 인자로 받은 l리스트의 원소 중에 중복 원소를 제거해주는 것이다.   
빈 리스트 s를 정의하고, for문과 if문을 이용해 l의 원소(e)가 s에 없으면 s에 e를 추가한다.
또한, l을 빈 리스트로 초기화할 때 l[:]를 사용하는 이유는, 그냥 l을 사용할 경우 self에 해당하는 \_\_str\_\_이 초기화되어버린다. 따라서 슬라이싱을 이용해 l을 list로 구분지어준 후에 빈 리스트로 초기화해야한다. 최종적으로 다시 빈 리스트인 l에 s 각각의 원소를 append 시켜준다.

## 문제 5

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    #@staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = [] #깊은 복사, 얕은 복사와 관련?
        for e in s:
            l.append(e)
    def __or__(self, right) : #합집합
        for e in right:
            if e in self :
                continue
            else :
                self.append(e)
        return self
    
    def __and__(self, right) : #교집합
        g = []
        for e in right :
            if e in self:
                g.append(e)
        self[:] = []
        for e in g:
            self.append(e)
        return self
    
    def __sub__(self, right) : #차집합
        s = self
        for e in right:
            if e in s:
                s.remove(e)
        return s

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    
    
    u = s | t
    print(u)
    
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    u = s & t
    print(u)
    
    s = MySet([1,2,3])
    t = MySet([3,4,5])
    u = s - t
    print(u)
```

4번 문제에서 추가된 것 - 합집합(|에 해당), 교집합(&에 해당), 차집합(-에 해당)  

합집합에 해당하는 \_\_or\_\_ 에서는 self, right 를 인자로 받고 이들은 각각 |문자 양쪽에 해당한다(ex. a | b). right에 해당하는 원소 e가 self안에 있으면 continue로 반복을 한 번 건너뛰고, 없을 경우 self에 추가한다. 최종적으로 self를 반환한다.  

교집합에 해당하는 \_\_and\_\_ 에서도 마찬가지로 같은것을 인자로 받고 형식또한 같다. 빈 리스트 g를 설정하고, right의 원소 e가 self안에도 있으면 g에 append 시킨다. 이후, self리스트를 빈 리스트로 수정하고, g에 해당하는 원소 e를 self에 추가한 뒤에 self를 반환한다.  

차집합에 해당하는 메소드는 \_\_sub\_\_ 이다. right의 원소 e가 self에 있으면 remove 메소드로 e를 없애고 s를 반환한다.

# NYtimes(http://www.nytimes.com)

```python
from urllib import request 
import string

url = 'http://www.nytimes.com'
url2 = 'http://edition.cnn.com'
url3 = 'http://www.foxnews.com'
url4 = 'http://www.chron.com'
url5 = 'http://www.latimes.com'
f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')
with open('www.nytimes.com.html','w',encoding='utf8') as f:
    f.write(source)
a = source
cnt=0
cnt
lists = []


while(cnt != len(a)) :
    
        if a[cnt] == '<' :
            if(a[cnt+1] == '!' and a[cnt+2]=='-') : #주석 삭제
                cnt+=3
                while True:
                    cnt+=1
                    if(a[cnt] == '-' and a[cnt+1] == '>'):
                        cnt+=1
                        break
            elif(a[cnt+1:cnt+7] == "script" ) : #스크립트 삭제
                cnt+=7
                while True:
                    cnt+=1
                    if(a[cnt:cnt+8] == "/script>"):
                        cnt+=8
                        break
                        
            elif(a[cnt+1:cnt+6]) == "style" : ##css 삭제
                cnt+=6
                while True:
                    cnt+=1
                    if(a[cnt:cnt+6]) == '/style':
                        cnt+=6
                        break
                        

                    
            else:
                while(a[cnt] != '>'): #HTML 삭제
                    cnt+=1
                    
        
        elif  a[cnt] != '<' and a[cnt] != '>' and a[cnt]!='\t' and a[cnt]!=' ' and a[cnt] !='\n' :
            start = cnt
            while a[cnt] != '<' and a[cnt] != '>' and a[cnt]!='\t' and a[cnt] !=' ' and a[cnt] !='\n' :
                    cnt+=1
                    if(cnt==len(a)):
                            break
                        
            end = cnt
            lists.append(a[start:end])
            
        else :
            cnt +=1
            
            
c=[]
for x in string.punctuation: #구두문자 리스트화
    c.append(x)
c.append("\r")
c.append("nbsp")
c.append("raquo")

for x in range(len(lists)) : #구두문자가 list의 원소안에 포함되어있을 경우
    for k in c:
        if k in lists[x]:
            lists[x] = lists[x].replace(k,'') #구두문자를 없앤다
    

lists.remove('') #구두문자만 있었을 경우 공백만 남으므로 공백 없애줌
            
dic = dict()
for x in lists:
    if x not in dic.keys(): #list의 원소가 사전에 없으면 추가
        dic[x] = 1
    elif x in dic.keys(): #이미 있는 원소면 1추가
        dic[x] += 1
        
f = open("StopwordList.txt","r")
stopword = f.readlines()
stopword_title = []
for i in range(len(stopword)):
    stopword[i] = stopword[i][:-1]
for i in range(len(stopword)):
    stopword_title.append(stopword[i].title())

dict_key_list = []
for i in dic.keys():
    dict_key_list.append(i)
for ele in stopword :
    if ele in dic.keys():
        del dic[ele]
for ele in stopword_title :
    if ele in dic.keys():
        del dic[ele]
while '' in dic.keys():
    del dic['']

print(dic)

FrequencyDic = dict()
FrequencyDic[url] = dic

    
```

```python
class OrderedWebWordsFrequency(WebWordsFrequency) :
    #Override
    def getWordsFrequency(self, reverse = True) :
        """ url = 'http://www.nytimes.com'
            url2 = 'http://edition.cnn.com'
            url3 = 'http://www.foxnews.com'
            url4 = 'http://www.chron.com'
            url5 = 'http://www.latimes.com' """
        if len(self.urls) == 0 :
            return {}
        
        self.comprehen = {} #종합본(단어 빈도수 사전)
        for url in FrequencyDic.keys():
            for ele in FrequencyDic[url].keys() :
                if ele in self.comprehen.keys():
                    self.comprehen[ele] += FrequencyDic[url][ele]
                else :
                    self.comprehen[ele] = FrequencyDic[url][ele]
        
        self.reverse = reverse
        if(self.reverse == False) :
            sorted_comprehen = sorted(self.comprehen.items(), key = lambda x : x[1], reverse = False)
        else :
            sorted_comprehen = sorted(self.comprehen.items(), key = lambda x : x[1], reverse = True)
        print(sorted_comprehen)
        
    def __iter__(self) : #반복자
        self.cnt =0
        return self
    def __next__(self) :
        if(self.cnt == len(self.urls)) :
            raise StopIteration
        result = self.urls[self.cnt]
        self.cnt +=1
        
        return result
    
w2 = OrderedWebWordsFrequency('http://www.nytimes.com','http://www.foxnews.com')

w2.getWordsFrequency()
for i in w2:
    print(i )
```

OrderedWebWordsFrequency 클래스는 WebWordsFrequency 를 상속받았다고는 하지만 위에서 계속 사용하던 w2와는 전혀다른 새로운 객체이다. 그러므로 단어를 종합받는 comprehen이라는 클래스멤버객체가 생성되있지 않으므로 위에서 했던것과 마찬가지로 그대로 추가해준다.
OrderedWebWordsFrequency 클래스의 getWordsFrequency 메소드는 기반 클래스의 메소드를 오버라이딩하는 것이므로 기능또한 다르게 구현된다.  
해당 메소드에는 self.reverse라는 객체가 있고, 여기에는 인자로 받은 reverse가 저장된다.
sort 메소드를 사용할 때 원래는 reverse를 False로주면 오름차순, True로 주면 내림차순이 되지만, reverse라는 자체 클래스 객체가 있기 때문에, 이에 의해 방식이 바뀔 수 있다. 이 코드에서는 기존과 정반대로 동작한다.

9번 - 

OrderedWebWordsFrequency 클래스 내부에 반복자 메소드인 \_\_iter\_\_ 메소드와 \_\_next\_\_메소드를 정의한다. \_\_iter\_\_은 반복자가 등장했을 경우 한번만 실행되어 self를 반환해준다. 이 때 \_\_next\_\_ 에서 사용될 self.cnt = 0 도 같이 정의해준다. 이는 self.urls 리스트의 인덱스에 사용되는 객체이다.  
\_\_next\_\_ 에서는 result에 self.urls를 인덱스마다 차례로 하나씩 대입해준다. 대입후에는 self.cnt를 1씩 증가시키고, self.cnt가 self.urls의 최대 인덱스에 도달할 시에 StopIteration 예외처리 시켜준다. 최종적으로 result를 반환하면 w2의 모든 원소(i)가 차례대로 출력된다.
